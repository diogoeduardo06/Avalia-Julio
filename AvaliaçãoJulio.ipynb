```python
uso_cpu = [
[20, 25, 40, 50, 45, 60, 55, 35, 70, 65],
[30, 35, 50, 60, 40, 55, 60, 45, 50, 55],
[15, 20, 30, 25, 35, 40, 45, 50, 55, 60],
[10, 15, 25, 35, 45, 50, 55, 60, 65, 70],
]
```

```python
def uso_maximo(matriz):
  for i in range(len(matriz[0])):
    soma = 0
    for row in matriz:
        soma += row[i]
    print(soma)
```

```python
uso_maximo(uso_cpu)
```

```python
def acima_da_media(matriz):
  for row in matriz:
    media = sum(row) / len(row)
    contagem = 0
    for i in row:
      if i > media:
        contagem += 1
    print(contagem)
    print(media)
```

```python
acima_da_media(uso_cpu)
```

```python
def dados_descrente(matriz):
    num_colunas = len(matriz[0])
    colunas_ordenadas = []

    for i in range(num_colunas):
        coluna = []
        for linha in matriz:
            coluna.append(linha[i])
        coluna.sort()
        colunas_ordenadas.append(coluna[::-1])
    print(colunas_ordenadas)
```

```python
dados_descrente(uso_cpu)
```
